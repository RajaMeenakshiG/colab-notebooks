```python
'''
✨ GROCERY LIST MANAGER ✨

1. Start with a basic grocery list: ["Milk", "Eggs", "Bread", "Butter", "Apples"]
2. Remove an item: "Bread"
3. Add a new priority item: "Sugar" at the beginning of the list
4. Alphabetically sort your list for better organization
5. Reverse the list to show most recently needed items first
6. Display the updated list and highlight the top 3 items you’ll buy first
'''

def grocery_list_manager():
    # Step 1: Create the initial list
    def create_list():
        return ["Milk", "Eggs", "Bread", "Butter", "Apples"]

    # Step 2–5: Update the list with given operations
    def update_list(grocery):
        print("Original List:", grocery)
        if "Bread" in grocery: # always check before removing
            grocery.remove("Bread")
            print('Removed "Bread" from the list.')
        grocery.insert(0, "Sugar")
        print('Inserted "Sugar" at the beginning.')
        grocery.sort()
        print("List after sorting:", grocery)
        grocery.reverse()
        print("List after reversing:", grocery)
        return grocery

    # Step 6: Display the final list and top 3 items
    def display_top_items(grocery):
        print("\n✅ Final Grocery List:", grocery)
        print("🛒 Top 3 Priority Items:", grocery[:3])

    # Execute all steps
    grocery = create_list()
    grocery = update_list(grocery)
    display_top_items(grocery)

grocery_list_manager()
```

```python
'''
🎓 STUDENT MARK ANALYSIS TOOL 🎓

1. Start with a set of student marks.
2. Add a newly received mark to the list.
3. Sort the marks for clarity.
4. Identify the top 3 performers (descending order).
5. Display detailed statistics:
   - Average mark (rounded to 2 decimals)
   - Highest mark
   - Lowest mark
'''

def analyze_student_marks():
    # Step 1: Initial student marks
    def fetch_initial_marks():
        return [96, 77, 100, 98, 99]

    # Step 2–3: Add new mark and sort the list
    def update_and_sort_marks(marks):
        new_mark = 97
        print(f"📥 Adding new mark: {new_mark}")
        marks.append(new_mark)
        marks.sort()
        print(f"📊 Sorted Marks: {marks}")
        return marks

    # Step 4–5: Show top 3 and other stats
    def display_statistics(marks):
        top_3 = sorted(marks, reverse=True)[:3]
        average = round(sum(marks) / len(marks), 2)
        highest = max(marks)
        lowest = min(marks)

        print("\n🔝 Top 3 Performers (Descending):", top_3)
        print(f"📈 Average Mark: {average}")
        print(f"🥇 Highest Mark: {highest}")
        print(f"🥉 Lowest Mark: {lowest}")

    # Execute all steps
    marks = fetch_initial_marks()
    marks = update_and_sort_marks(marks)
    display_statistics(marks)

analyze_student_marks()
```

```python
'''
✅ SMART TO-DO TASK MANAGER ✅

1. Create your initial to-do list.
2. Add new tasks to your planner.
3. Remove tasks marked as completed.
4. Display pending tasks in reverse — starting with the most recent!
'''

def todo_task_manager():
    # Step 1: Create an initial list of tasks
    def fetch_initial_tasks():
        return ["Send Email", "Complete Report", "Buy Groceries", "Call Mentor"]

    # Step 2–3: Update the list with new tasks and remove completed ones
    def update_task_list(task_list):
        new_tasks = ["Pay Bills", "Water Plants"]
        completed_tasks = ["Send Email", "Pay Bills"]

        print("📝 Initial Tasks:", task_list)
        print("➕ Adding Tasks:", new_tasks)
        task_list.extend(new_tasks)

        print("✅ Completed Tasks:", completed_tasks)
        for task in completed_tasks:
            if task in task_list:
                task_list.remove(task)

        task_list.reverse()
        return task_list

    # Step 4: Display pending tasks in reverse order
    def show_pending_tasks(pending):
        print("\n🕒 Pending Tasks (Most Recent First):")
        for idx, task in enumerate(pending, start=1):
            print(f"{idx}. {task}")

    # Execute all steps
    tasks = fetch_initial_tasks()
    updated_tasks = update_task_list(tasks)
    show_pending_tasks(updated_tasks)

# Launch the manager
todo_task_manager()
```

```python
'''
📦 ONLINE ORDER TRACKING SYSTEM 🚚

Operations:
1. Initialize the day's order queue.
2. Track delivered and pending orders.
3. Cancel an unwanted order and add a new one.
4. Perform end-of-day cleanup.
'''

def online_order_tracker():
    # Step 1: Create today's initial order list
    def initialize_orders():
        orders = ["Order#101", "Order#102", "Order#103", "Order#104", "Order#105"]
        print("🗂️ Initial Orders for Today:")
        print(", ".join(orders), "\n")
        return orders

    # Step 2: Update orders - cancel, add new, track status
    def process_orders(orders):
        delivered = orders[:3]           # first 3 delivered
        pending = orders[3:]             # rest pending

        cancel_order = "Order#105"
        new_order = "Order#106"

        print(f"❌ Cancelling {cancel_order}...")
        if cancel_order in orders:
            orders.remove(cancel_order)
            print(f"   - {cancel_order} successfully removed.")
        else:
            print(f"   - {cancel_order} not found.")

        print(f"➕ Adding New Order: {new_order}")
        orders.append(new_order)

        return orders, delivered, pending

    # Step 3: Display current status and clear for next day
    def display_summary(current_orders, delivered, pending):
        print("\n✅ Delivered Orders:", ", ".join(delivered))
        print("⏳ Still Pending:", ", ".join(pending))
        print("📋 Updated Order Queue:", ", ".join(current_orders))

        print("\n🧹 Performing End-of-Day Cleanup...")
        current_orders.clear()
        print("📭 All orders cleared for tomorrow:", current_orders)

    # Execute all steps
    order_list = initialize_orders()
    updated_orders, delivered_orders, pending_orders = process_orders(order_list)
    display_summary(updated_orders, delivered_orders, pending_orders)

online_order_tracker()
```

```python
'''
🧠 MEMORY FLIP GAME – TEST YOUR MEMORY!

Goal: Flip two hidden cards. If they match, you earn a point.

Features:
1. Setup hidden pairs.
2. Flip two positions randomly.
3. Match? Score a point.
4. Play 7 rounds and reveal final score.

Flip wisely and have fun! 🎯
'''

import random

def memory_flip_game():
    score = 0  # Track total points

    # Step 1: Prepare the card set
    def setup_game():
        items = ["🦚", "🦋", "🦚", "🦋"]
        random.shuffle(items)
        return items

    # Step 2: Flip and check two cards
    def flip_and_check(items, round_num):
        nonlocal score
        i, j = random.sample(range(len(items)), 2)
        print(f"🎲 Round {round_num}:")
        print(f"   🔄 You flipped cards at positions {i} and {j}")
        print(f"   🃏 Revealed: {items[i]} and {items[j]}")
        if items[i] == items[j]:
            print("   ✅ It's a match!")
            score += 1
        else:
            print("   ❌ Not a match. Try again!")
        print()

    # Step 3: Run the game for 7 rounds
    for round_num in range(1, 8):
        cards = setup_game()
        flip_and_check(cards, round_num)

    # Step 4: Final Score
    print(f"🏁 Game Over! You scored {score} out of 7.")
    if score == 7:
        print("🌟 Perfect Memory! You're a memory master!")
    elif score >= 4:
        print("👏 Great job! Keep training your brain!")
    else:
        print("🔁 Try again! Memory needs some work.")

# Start the game
memory_flip_game()
```

```python
'''
🎯 HANGMAN – GUESS THE SECRET WORD!

Test your word-guessing skills before the stickman is fully drawn!

Features:
1. Pick a secret 5-letter word.
2. Guess one letter at a time.
3. Reveal letters or lose a life.
4. Survive 7 wrong guesses to win!

Hang in there! 🪢
'''

import random

def hangman_game():
    def get_clues(length):
        return ["❓"] * length

    def get_word():
        words = ["robot", "angel", "teddy", "magic", "phone", "music", "cloud"]
        return random.choice(words)

    def get_stickman(stage):
        stages = [
            """




            """,
            """
                |
                |
                |
                |
                |_____
            """,
            """
                |------
                |
                |
                |
                |
                |_____
            """,
            """
                |------
                |    O
                |
                |
                |
                |_____
            """,
            """
                |------
                |    O
                |    |
                |
                |
                |_____
            """,
            """
                |------
                |    O
                |   /|
                |
                |
                |_____
            """,
            """
                |------
                |    O
                |   /|\\
                |
                |
                |_____
            """,
            """
                |------
                |    O
                |   /|\\
                |   / \\
                |
                |_____
            """
        ]
        return stages[stage]

    def update_clues(guess, secret, clues):
        for i, ch in enumerate(secret):
            if ch == guess:
                clues[i] = guess

    def is_incomplete(clues):
        return "❓" in clues

    # Game Start
    secret_word = get_word()
    clues = get_clues(len(secret_word))
    attempts = 0
    max_attempts = 7

    print("🎮 Welcome to Hangman – Tamil Edition!")
    print("💡 Guess the secret 5-letter English word.\n")

    while attempts < max_attempts:
        print("🧩 Current Clues:", " ".join(clues))
        print(f"❤️ Lives Left: {max_attempts - attempts}")
        guess = input("🔡 Enter a letter: ").lower()

        if not guess.isalpha() or len(guess) != 1:
            print("⚠️ Invalid input. Enter a single letter.\n")
            continue

        if guess in secret_word and guess not in clues:
            print(f"✅ Good job! '{guess}' is in the word.\n")
            update_clues(guess, secret_word, clues)
        elif guess in clues:
            print(f"🔁 You already guessed '{guess}'. Try something new!\n")
        else:
            print(f"❌ Oops! '{guess}' is not in the word.")
            attempts += 1
            print(get_stickman(attempts))
            print()

        if not is_incomplete(clues):
            print(f"\n🎉 Congrats! You found the word: {''.join(clues)}")
            break

    if attempts == max_attempts:
        print(f"\n💀 Game Over! The word was: {secret_word.upper()}")

hangman_game()
```

```python
'''
🏏 IPL AUCTION SYSTEM

Be the franchise boss! Bid on your favorite players and win them for your team.

Features:
1. View players with base price & current bids.
2. Add/remove players from auction.
3. Place bids and compete for superstars.
4. Display final auction results.

Let the bidding war begin! 💰
'''

import random

def ipl_auction():
    players = ['MS Dhoni', 'Virat Kohli', 'Rohit Sharma', 'Suresh Raina']
    base_prices = [150, 140, 120, 100] # in millions
    bids = base_prices.copy()

    def show_players():
        if not players:
            print("\n❗ No players available for auction.")
        else:
            print("\n📋 Available Players:")
            for idx in range(len(players)):
                print(f"{idx + 1}. 🧢 {players[idx]} | Base: ₹{base_prices[idx]}M | Bid: ₹{bids[idx]}M")

    def add_player():
        name = input("➕ Enter player name: ").strip()
        base = input("💵 Enter base price (in millions): ").strip()
        if base.isdigit():
            base = int(base)
            players.append(name)
            base_prices.append(base)
            bids.append(base)
            print(f"✅ {name} added with base price ₹{base}M")
        else:
            print("❌ Invalid base price. Must be a number.")

    def remove_player():
        show_players()
        idx = input("❌ Enter player number to remove: ").strip()
        if idx.isdigit():
            idx = int(idx)
            if 1 <= idx <= len(players):
                removed = players.pop(idx - 1)
                base_prices.pop(idx - 1)
                bids.pop(idx - 1)
                print(f"🗑️ Removed player: {removed}")
            else:
                print("⚠️ Invalid number.")
        else:
            print("❌ Please enter a valid number.")

    def bid_player():
        show_players()
        idx = input("🎯 Enter player number to bid on: ").strip()
        if idx.isdigit():
            idx = int(idx)
            if 1 <= idx <= len(players):
                name = input("🧍 Your name: ").strip()
                bid = input("💰 Enter your bid (in millions): ").strip()
                if bid.isdigit():
                    bid = int(bid)
                    if bid > bids[idx - 1]:
                        bids[idx - 1] = bid
                        print(f"👏 {name} placed ₹{bid}M on {players[idx - 1]}")
                    else:
                        print(f"🚫 Your bid is too low. Current bid: ₹{bids[idx - 1]}")
                else:
                    print("❌ Invalid bid amount.")
            else:
                print("⚠️ Invalid player number.")
        else:
            print("❌ Please enter a valid number.")

    def show_results():
        if not players:
            print("\n📭 No auction results yet.")
        else:
            print("\n🏆 Final Auction Results:")
            for i in range(len(players)):
                print(f"🔹 {players[i]} – ₹{bids[i]}M")

    # Auction Menu
    while True:
        print("\n🎙️ IPL Mega Auction – Menu")
        print("1️⃣ Show Available Players")
        print("2️⃣ Add a New Player")
        print("3️⃣ Remove a Player")
        print("4️⃣ Place a Bid")
        print("5️⃣ Show Auction Results")
        print("6️⃣ Exit Auction")

        choice = input("👉 Enter your choice: ").strip()

        if choice == '1':
            show_players()
        elif choice == '2':
            add_player()
        elif choice == '3':
            remove_player()
        elif choice == '4':
            bid_player()
        elif choice == '5':
            show_results()
        elif choice == '6':
            print("\n🚪 Exiting Auction. Thanks for playing!")
            break
        else:
            print("⚠️ Invalid option. Try again.")

# Start the auction
ipl_auction()
```
